```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage=requests.get('http://books.toscrape.com/',headers=headers).text
```

```python
soup=BeautifulSoup(webpage,'html.parser')
```

```python
print(soup.prettify())
```

```python
names = []
for img in soup.find_all('img'):
    alt_text = img.get('alt')
    if alt_text:
        names.append(alt_text.strip())
```

```python
names
```

```python
price=[]
for i in soup.find_all('p',class_='price_color'):
  price.append(i.text.strip())

for i in range(len(price)):
  price[i]=price[i][1:]
```

```python
price
```

```python
stock=[]
for i in soup.find_all('p',class_='instock availability'):
  stock.append(i.text.strip())
```

```python
stars = []
for tag in soup.find_all('p', class_=lambda x: x and 'star-rating' in x):
    stars.append(tag.get('class')[1])
```

```python
stars
```

```python
d={'Book Name':names,'Price':price,'Stock Availability':stock,"Rating":stars}
df=pd.DataFrame(d)
df
```
